```python
import matplotlib.pyplot as plt
print(2.1)
# Data
people = ["Kiran", "Arun", "Vijay", "Varun"]
ages = [25, 30, 35, 40]
heights = [145, 151, 165, 173]
weights = [45, 55, 65, 75]

# Create figure and axis objects
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Plot age
axs[0].bar(people, ages, color='skyblue')
axs[0].set_title('Age of People')
axs[0].set_xlabel('People')
axs[0].set_ylabel('Age')

# Plot height
axs[1].bar(people, heights, color='lightgreen')
axs[1].set_title('Height of People')
axs[1].set_xlabel('People')
axs[1].set_ylabel('Height (cm)')

# Plot weight
axs[2].bar(people, weights, color='salmon')
axs[2].set_title('Weight of People')
axs[2].set_xlabel('People')
axs[2].set_ylabel('Weight (kg)')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
print(2.2)
# Data
people = ["Kiran", "Arun", "Vijay", "Varun"]
ages = [25, 30, 35, 40]
heights = [145, 151, 165, 173]
weights = [45, 55, 65, 75]

# Create a DataFrame for easier plotting with seaborn
df = pd.DataFrame({
    'Person': people,
    'Age': ages,
    'Height': heights,
    'Weight': weights
})

# Plotting
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Scatter plot: Age vs Height
axs[0].scatter(df['Age'], df['Height'], color='blue')
axs[0].set_title('Age vs Height')
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Height (cm)')
for i, txt in enumerate(df['Person']):
    axs[0].annotate(txt, (df['Age'][i], df['Height'][i]))

# Scatter plot: Age vs Weight
axs[1].scatter(df['Age'], df['Weight'], color='green')
axs[1].set_title('Age vs Weight')
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Weight (kg)')
for i, txt in enumerate(df['Person']):
    axs[1].annotate(txt, (df['Age'][i], df['Weight'][i]))

# Scatter plot: Height vs Weight
axs[2].scatter(df['Height'], df['Weight'], color='red')
axs[2].set_title('Height vs Weight')
axs[2].set_xlabel('Height (cm)')
axs[2].set_ylabel('Weight (kg)')
for i, txt in enumerate(df['Person']):
    axs[2].annotate(txt, (df['Height'][i], df['Weight'][i]))

plt.tight_layout()
plt.show()

# Calculate and display correlation matrix
correlation_matrix = df[['Age', 'Height', 'Weight']].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Pair Plot
sns.pairplot(df[['Age', 'Height', 'Weight']], kind='scatter', diag_kind='kde')
plt.show()
```
