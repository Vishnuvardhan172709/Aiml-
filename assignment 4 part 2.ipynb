```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

np.random.seed(42)
X = np.random.rand(100, 1)
Y = 3 * X + 2 + np.random.randn(100, 1) * 0.5

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

model = LinearRegression()

model.fit(X_train, Y_train)

intercept = model.intercept_
slope = model.coef_

print(f"Intercept (b): {intercept[0]}")
print(f"Slope (m): {slope[0][0]}")

Y_pred = model.predict(X_test)
mse = mean_squared_error(Y_test, Y_pred)
r_squared = model.score(X_test, Y_test)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r_squared}")

sorted_pred = np.sort(Y_pred.flatten())
cdf = np.arange(1, len(sorted_pred) + 1) / len(sorted_pred)

plt.figure(figsize=(8, 6))
plt.plot(sorted_pred, cdf, marker='.', linestyle='none')
plt.title("Cumulative Distribution Function (CDF) of Predicted Values")
plt.xlabel("Predicted Y values")
plt.ylabel("CDF")
plt.grid(True)
plt.show()
```
